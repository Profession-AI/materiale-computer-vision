### **RGB Image**

```python
plt.imshow(image) # che Ã¨ successo?
```

```python
# Load the image in color (BGR format)
image_bgr = cv2.imread('rgb.jpg')

# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the original and converted images
plt.subplot(1, 2, 1)
plt.title('BGR Image')
plt.imshow(image_bgr)  # Convert BGR to RGB for correct display with matplotlib

plt.subplot(1, 2, 2)
plt.title('RGB Image')
plt.imshow(image_rgb)

plt.show()
```
